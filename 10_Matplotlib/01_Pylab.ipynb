Pylab combina funcionalidades da biblioteca Matplotlib com pacote Numpy.

```python
import matplotlib as mpl
from pylab import *
#garantindo que os gráficos serão criados dentro do jupyter
%matplotlib inline
```

# Gráfico de Linha

```python
#Dados
x = linspace(0, 5, 10)
y = x ** 2

#Cria a Figura
fig = plt.figure()

#Escala dos Eixos
axes = fig.add_axes([0, 0, 0.8, 0.8])

#Plot
axes.plot(x, y, 'r')

#Labels e Título
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Grafico de Linha');
```

```python
#Gráfico de linha com 2 figuras

x = linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

axes1 = fig.add_axes ([0.1, 0.1, 0.8, 0.8]) # eixos da figura principal
axes2 = fig.add_axes ([0.2, 0.5, 0.4, 0.3]) # eixos da figura secundária
                       
# Figura principal
axes1.plot(x, y,'r') 
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura Principal')
                       
# Figura secundária 
axes2.plot(y, x,'g')
axes2.set_xlabel( 'y')
axes2.set_ylabel ('x')
axes2.set_title( 'Figura Secundária');
```

```python
# Gráficos de linha em Paralelo

# Dados
x = linspace(0, 5, 10)
y = x ** 2

# Divira a área de plotagem em dois subplots
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# Loop pelos eixos para criar cada plot
for ax in axes: 
    ax.plot(x, y, 'r') 
    ax.set_xlabel ('x')
    ax.set_ylabel('y')
    ax.set_title('Título')

# Ajusta o layout
fig.tight_layout ()
```

```python
# Gráficos de linha com diferentes escalas

# Dados
x = linspace(0, 5, 10)
y = x ** 2

# Cria os subplots
fig, axes = plt.subplots(1, 2, figsize=(10,4))

# Cria o plotl
axes[0].plot(x, x**2, x, exp(x))
axes[0].set_title("Escala Padrão")

# Cria o plot2
axes[1].plot (x, x**2, x, exp (x))
axes[1].set_yscale( "log" ) 
axes [1].set_title("Escala Logaritmica(y)");
```

```python
# Grid

# Dados
x = linspace(0, 5, 10)
y = x ** 2

# Cria os subplots
fig, axes = plt.subplots(1, 2, figsize = (10,3))

# Grid padrão
axes[0].plot(x, x**2, x, x**3, lw = 2)
axes[0].grid (True)

# Grid customizado
axes[1].plot(x, x**2, x, x**3, lw = 2)
axes[1].grid(color = 'b', alpha = 0.7, linestyle = 'dashed', linewidth = 0.8)
```

```python
# Diferentes estilos de Plots

# Dados
xx = np. linspace(-0.75, 1., 100)
n = np. array ([0,1,2,3,4,5])

# Subplots
fig, axes = plt.subplots(1, 4, figsize = (12,3))

#Plot 1
axes[0].scatter (xx, xx + 0.25 * randn(len(xx)), color = "black" )
axes[0].set_title("scatter")

# Plot 2
axes[1].step(n, n ** 2, lw = 2, color = "blue")
axes[1].set_title("step")

# Plot 3
axes[2].bar(n, n ** 2, align = "center", width = 0.5, alpha = 0.5, color = "magenta")
axes[2].set_title("step")

#Plot 4
axes [3].fill_between(x, x ** 2, x ** 3, alpha = 0.5, color = "green");
axes [3].set_title("fill_between");
```
